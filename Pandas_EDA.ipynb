```python
numerical_df = df.select_dtypes(include = np.number).corr()
numerical_df
```

```python
plt.figure(figsize=(20, 7))
sns.heatmap(numerical_df, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Numerical Features')
plt.show()
```

```python
df.groupby('Continent').mean(numeric_only=True).sort_values(by='2022 Population',ascending=False)
```

```python
df['Total_Population'] = df['2022 Population'] + df['2020 Population'] + df['2015 Population']
```

```python
df_transpose = df.groupby('Continent').mean(numeric_only = True).sort_values(by = 'Total_Population',ascending = False)
df_transpose
```
